```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
df = pd.read_csv("PCA_lat_lon.csv")
locations = df['city_name']
```

```python
results = {}
for loc in locations:
    res = requests.get(f"https://unsplash.com/s/photos/{loc.lower().replace(' ','-')}")
    soup = bs(res.content)
    imgs = soup.find_all("img", class_="tB6UZ a5VGX")
    try:
        results[loc] = [i['src'] for i in imgs if 'premium' not in i['src']][0]
    except:
        results[loc] = "https://images.unsplash.com/photo-1523595857-fe9ee689f76f?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxzZWFyY2h8MTR8fGNhbGlmb3JuaWF8ZW58MHx8MHx8&auto=format&fit=crop&w=500&q=60"
```

```python
import json
with open('location_image.json', 'w') as f:
    json.dump(results, f)
```
